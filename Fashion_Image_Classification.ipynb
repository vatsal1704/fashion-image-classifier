```python
# Load the dataset
from tensorflow.keras.datasets import fashion_mnist
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
(x_train,y_train),(x_test,y_test)=fashion_mnist.load_data()
```

```python
print('Shape of x_train:',x_train.shape)
print('Shape of x_test:',x_test.shape)
print('Shape of y_train:',y_train.shape)
print('Shape of x_test:',x_test.shape)
```

```python
x_train[0]
```

```python
# the data is 60000 images of pixel size (28,28)
# Min and Max values in pixel
print('Min Pixel Value:',x_train.min())
print('Max Pixel Value:',x_train.max())
```

```python
# Unique labels
print(np.unique(y_train))
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# let's get some images along with their labels
nrows=4
ncols=10
t=1
plt.figure(figsize=(15,6))
for i in range(nrows*ncols):
  plt.subplot(nrows,ncols,t)
  plt.imshow(x_train[i],cmap='binary',interpolation='nearest')
  plt.title(f'{class_names[y_train[i]]}',color='red',fontsize=8)
  t+=1
  plt.axis('off')
plt.subplots_adjust(hspace=0.2,wspace=0.2)
plt.show()

```

```python
# Image is a 3D data we need to make it 2D
# Normalising the pixels and Reshaping the image
# Normalisation 0-255 ---> 0-1 (divide pixels by 255) (for faster computation)
# Reshaping   60000 X 28 X 28 ---->60000 X 784



#Normalise the pixels
x_train=x_train/255
x_test=x_test/255
```

```python
# reshape the data
x_train_s=x_train.reshape(x_train.shape[0],28*28)
x_test_s=x_test.reshape(x_test.shape[0],28*28)
```

```python
x_train_s.shape
```

```python
x_test.shape
```

## Build a NN Model


```python
# build a neural net model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
# create a sequential model
model=Sequential()


# add input layer and first hidden layer in sequential model
model.add(Dense(128,input_shape=(784,),activation='relu'))

# lets add another layer
model.add(Dense(64,activation='relu'))

# lets add output layer
model.add(Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
# we need to compile the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train_s,y_train,epochs=60,batch_size=10000)
```

```python
plt.plot(model.history.history['loss'])
plt.plot(model.history.history['accuracy'])
plt.show()
```

```python
# lets evaluate this model
loss,acc=model.evaluate(x_test_s,y_test)
```

```python
print(f'Accuracy:{acc}')
print(f'Loss:{loss}')
```

```python
# lets get the predictions
prediction=model.predict(x_test_s)
```

```python
pred_labels=np.argmax(prediction,axis=1)
```

```python
pred_labels
```

```python
# Parameters
nrows, ncols = 4, 10
num_images = nrows * ncols

plt.figure(figsize=(18, 7))  # Wider view for better spacing

for i in range(num_images):
    plt.subplot(nrows, ncols, i + 1)
    plt.imshow(x_test[i], cmap="binary", interpolation="nearest")

    actual_class = class_names[y_test[i]]
    pred_class = class_names[pred_labels[i]]

    # Color coding: blue if correct, red if incorrect
    label_color = "blue" if actual_class == pred_class else "red"

    # Set title as a short label
    plt.title(f"{actual_class}  {pred_class}", fontsize=8, color=label_color)

    # Optionally add xlabel (usually small and not necessary if title is enough)
    plt.xlabel(f"True:{actual_class}\nPred:{pred_class}", fontsize=6, color=label_color)

    plt.axis("off")

# Adjust spacing between plots
plt.subplots_adjust(hspace=1.4, wspace=0.4)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred_labels))
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(confusion_matrix(y_test,pred_labels),annot=True,cmap='RdBu',fmt='g')
plt.xticks(range(10),class_names,rotation=90)
plt.yticks(range(10),class_names,rotation='horizontal')
plt.show()
```
